### Scrape land price data from http://www.land.mlit.go.jp/landPrice/SearchServlet?MOD=2&nen_check=new&tounenkakaku_from=&tounenkakaku_to=&MOD=2&TDK=&SKC=13103&CHI=&YFR=2019&YTO=2019&YOU=&PFR=&PTO=&PG=0&LATEST_YEAR=1&sp=true

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.land.mlit.go.jp/landPrice/SearchServlet"
url += "?MOD=2"
url += "&nen_from=1990&nen_from=2019"
url += "&tounenkakaku_from=&tounenkakaku_to="
url += "&MOD=2&TDK=&SKC=13123"
url += "&CHI=&YFR=1990&YTO=2019&YOU=&PFR=&PTO=&PG=0&LATEST_YEAR=0"
```

```python
html = requests.get(url).text
soup = BeautifulSoup(html)
```

```python
result = []
for datalist in soup.find_all("div", attrs={"class": "datalist"}):
    address = datalist.find_all("div", attrs={"class": "datavalue2"})[0].text.split("\u3000")[0]
    price = datalist.find_all("div", attrs={"class": "datavalue"})[2].text.replace("(円/m²)", "").replace(",", "")
    station = datalist.find_all("div", attrs={"class": "datavalue"})[3].text.split("、")[0]
    duration =datalist.find_all("div", attrs={"class": "datavalue"})[3].text.split("、")[1].replace("\t","").replace(" ","").replace("m","").replace(",", "")
    area = datalist.find_all("div", attrs={"class": "datavalue"})[4].text.replace("(m²)", "")
    
    result.append({
        "address": address,
        "price": price,
        "station": station,
        "duration": duration,
        "area": area,
    })
    
```

```python
pd.DataFrame(result)
```
